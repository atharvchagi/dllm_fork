```python
import re

# Count how many times "we" appears as a standalone word in generation text
def _flatten_text(x):
    if isinstance(x, str):
        return x
    if isinstance(x, list):
        return " ".join(_flatten_text(v) for v in x)
    if isinstance(x, dict):
        return " ".join(_flatten_text(v) for v in x.values())
    return str(x)

df["we_count"] = df["resps"].apply(
    lambda x: len(re.findall(r"\bwe\b", _flatten_text(x).lower()))
)
```
